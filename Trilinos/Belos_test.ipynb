```python
import numpy as np
import scipy as sp
import scipy.io as sio
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
A = sio.mmread('../build/A_TCMAT.mtx').toarray()
b = sio.mmread('../build/b_TV.mtx')
xTrue = sio.mmread('../build/xTrue_TV.mtx')
x = sio.mmread('../build/x_TV.mtx')
print(np.linalg.cond(A))
fig, ax = plt.subplots()
from matplotlib.colors import LogNorm
cs = ax.imshow(A,norm=LogNorm())
cbar = fig.colorbar(cs)
```

```python
xsolGMRES = sio.mmread('../build/xsolGMRES_TV.mtx')
xsolBICGSTAB = sio.mmread('../build/xsolBICGSTAB_TV.mtx')
```

```python
xErrorInit = x - xTrue
xErrorGMRES=xsolGMRES-xTrue
bErrorGMRES = np.dot(A,xsolGMRES) - b
xErrorBICGSTAB=xsolBICGSTAB-xTrue
bErrorBICGSTAB = np.dot(A,xsolBICGSTAB) - b
```

```python
xErrorInitNorm = np.linalg.norm(xErrorInit)
xErrorGMRESNorm=np.linalg.norm(xErrorGMRES)
bErrorGMRESNorm = np.linalg.norm(bErrorGMRES) 
xErrorBICGSTABNorm=np.linalg.norm(xErrorBICGSTAB)
bErrorBICGSTABNorm = np.linalg.norm(bErrorBICGSTAB) 

print(xErrorInitNorm)
print(xErrorGMRESNorm,bErrorGMRESNorm)
print(xErrorBICGSTABNorm,bErrorBICGSTABNorm)
```
